```python
# 독립된 창에서 띄우기 (안되면 %matplotlib tk 로 시도)
%matplotlib qt 

import numpy as np
import matplotlib.pyplot as plt
import time

# 1. 데이터 생성
np.random.seed(42)
X = 2 * np.random.rand(50, 1)
y = 4 + 3 * X + np.random.randn(50, 1)
X_b = np.c_[np.ones((50, 1)), X]

# 2. 그래프 초기 설정 (루프 밖에서 한 번만 생성!)
plt.ion() # 대화형 모드 활성화
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 손실 함수 곡선(초록색 U자)은 변하지 않으므로 미리 계산
w_range = np.linspace(-1, 7, 100)
l_range = [np.mean((X * w + 4 - y)**2) for w in w_range]

try:
    while True:
        # 매 사이클마다 파라미터 초기화
        theta = np.array([[np.random.uniform(0, 8)], [np.random.uniform(0, 6)]])
        theta_path, loss_path = [], []
        
        for i in range(30):
            # 경사하강법 계산
            predictions = X_b.dot(theta)
            gradients = 2/50 * X_b.T.dot(predictions - y)
            theta = theta - 0.1 * gradients
            
            loss = np.mean((predictions - y)**2)
            theta_path.append(theta[1][0])
            loss_path.append(loss)
            
            # --- 기존 창의 내용만 지우고 다시 그리기 ---
            ax1.cla() # Clear axis 1
            ax1.scatter(X, y, color='red', alpha=0.5)
            ax1.plot(X, X_b.dot(theta), color='blue', linewidth=3)
            ax1.set_title(f"Model Fitting (Step {i+1})")
            ax1.set_ylim(0, 15)
            
            ax2.cla() # Clear axis 2
            ax2.plot(w_range, l_range, 'g-', alpha=0.3)
            ax2.plot(theta_path, loss_path, 'ro-', markersize=6)
            ax2.set_title("Loss Function Space")
            ax2.set_xlabel("Weight (w)")
            
            # 창을 새로 띄우지 않고 현재 창을 갱신
            plt.draw()
            plt.pause(0.05) # 애니메이션 속도
            
        time.sleep(1.5) # 한 세션 끝나면 잠시 대기

except KeyboardInterrupt:
    plt.ioff()
    print("사용자에 의해 종료되었습니다.")
```

```python
%matplotlib qt 

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time

# 1. 데이터 생성
np.random.seed(42)
X = 2 * np.random.rand(50, 1)
y = 4 + 3 * X + np.random.randn(50, 1)
X_b = np.c_[np.ones((50, 1)), X]

# 2. 3D 손실 함수 지형 계산 (그릇 모양)
w_range = np.linspace(0, 6, 40) # 기울기 범위
b_range = np.linspace(0, 8, 40) # 절편 범위
W, B = np.meshgrid(w_range, b_range)
Z = np.array([np.mean((X * w + b - y)**2) for w, b in zip(np.ravel(W), np.ravel(B))])
Z = Z.reshape(W.shape)

# 3. 시각화 설정
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
plt.ion()

try:
    while True:
        # 초기 파라미터 랜덤 설정
        theta = np.array([[np.random.uniform(0, 8)], [np.random.uniform(0, 6)]])
        path_w, path_b, path_loss = [], [], []

        for i in range(30):
            # 경사하강법
            predictions = X_b.dot(theta)
            gradients = 2/50 * X_b.T.dot(predictions - y)
            theta = theta - 0.1 * gradients # lr=0.1
            
            loss = np.mean((predictions - y)**2)
            path_b.append(theta[0][0])
            path_w.append(theta[1][0])
            path_loss.append(loss)

            # 그래프 업데이트
            ax.clear()
            # 초록색 오차 곡면 (지도)
            ax.plot_surface(W, B, Z, alpha=0.4, cmap='viridis') 
            # 빨간색 경로 (지도 위에 딱 붙어서 이동!)
            ax.plot(path_w, path_b, path_loss, 'ro-', markersize=5, linewidth=2)
            ax.scatter(path_w[-1], path_b[-1], path_loss[-1], c='red', s=100) # 현재 위치
            
            ax.set_title(f"Step {i+1}: Descent on Loss Surface")
            ax.set_xlabel("Weight (w)")
            ax.set_ylabel("Bias (b)")
            ax.set_zlabel("Loss")
            
            plt.draw()
            plt.pause(0.1)
            
        time.sleep(2)

except KeyboardInterrupt:
    plt.ioff()
    print("종료되었습니다.")
```

```python
%matplotlib qt 

import numpy as np
import matplotlib.pyplot as plt
import time

# 1. 데이터 생성 (y = 3x + 4 형태)
np.random.seed(42)
X = 2 * np.random.rand(50, 1)
# 이번에는 절편을 4로 완전히 고정하고 학습할 겁니다.
y = 4 + 3 * X + np.random.randn(50, 1)

# 2. 그래프 초기 설정
plt.ion()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 오차 지도(초록색 선) 미리 계산 (절편 b=4로 고정된 상태)
w_range = np.linspace(-1, 7, 100)
l_range = [np.mean((X * w + 4 - y)**2) for w in w_range]

try:
    while True:
        # 기울기(w)만 랜덤하게 시작 (절편은 4로 고정)
        w = np.random.uniform(0, 6)
        w_path, loss_path = [], []
        
        for i in range(30):
            # 예측값 (y_hat = w * X + 4)
            prediction = X * w + 4
            
            # 오차 계산 및 미분 (MSE를 w에 대해서만 미분)
            error = prediction - y
            gradient = 2 * np.mean(error * X)
            
            # 경사하강법으로 w 업데이트
            w = w - 0.1 * gradient # lr=0.1
            
            loss = np.mean(error**2)
            w_path.append(w)
            loss_path.append(loss)
            
            # --- 그래프 업데이트 ---
            ax1.cla()
            ax1.scatter(X, y, color='red', alpha=0.5, label='Data')
            ax1.plot(X, X * w + 4, color='blue', linewidth=3, label='Model')
            ax1.set_title(f"Model Space (Step {i+1})")
            ax1.set_ylim(0, 15)
            
            ax2.cla()
            # 초록색 지도 그리기
            ax2.plot(w_range, l_range, 'g-', alpha=0.5, label='Loss Curve')
            # 빨간 점을 지도(선) 위에 찍기
            ax2.plot(w_path, loss_path, 'ro-', markersize=6, alpha=0.8)
            ax2.scatter(w, loss, color='red', s=100, edgecolors='black', zorder=5)
            ax2.set_title("Loss Space (On the line!)")
            ax2.set_xlabel("Weight (w)")
            ax2.set_ylabel("Loss")
            
            plt.draw()
            plt.pause(0.1)
            
        time.sleep(1.5)

except KeyboardInterrupt:
    plt.ioff()
    print("시뮬레이션을 종료합니다.")
```

```python
%matplotlib qt

import numpy as np
import matplotlib.pyplot as plt
import time

# 1. 가상 데이터 생성 (정답: y = 3x + 4)
np.random.seed(42)
X = 2 * np.random.rand(50, 1)
y = 4 + 3 * X + np.random.randn(50, 1)

# 2. 고정된 오차 지도(초록색 선) 데이터 미리 계산
w_range = np.linspace(-1, 7, 100)
l_range = [np.mean((X * w + 4 - y)**2) for w in w_range]

# 3. 4개의 단계별 그래프 설정
plt.ion()
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
plt.subplots_adjust(hspace=0.3, wspace=0.2)

try:
    while True:
        # 무작위 기울기에서 시작
        w = np.random.uniform(0, 6)
        w_path, loss_path = [], []
        
        for i in range(25):
            # 현재 모델 상태 계산
            prediction = X * w + 4
            error = prediction - y
            loss = np.mean(error**2)
            gradient = 2 * np.mean(error * X) # 미분값(기울기)
            
            w_path.append(w)
            loss_path.append(loss)

            # --- [1번 그래프] 데이터와 현재 모델 ---
            axes[0, 0].cla()
            axes[0, 0].scatter(X, y, color='red', alpha=0.5)
            axes[0, 0].plot(X, prediction, color='blue', linewidth=3)
            axes[0, 0].set_title("Step 1: Current Model (Line)")
            axes[0, 0].set_ylim(0, 15)
            axes[0, 0].set_ylabel("Target (y)")

            # --- [2번 그래프] 실제 오차(잔차) 시각화 ---
            axes[0, 1].cla()
            axes[0, 1].scatter(X, y, color='red', alpha=0.5)
            axes[0, 1].plot(X, prediction, color='blue', linewidth=2)
            # 수직 점선으로 (y - y_hat) 시각화
            for xi, yi, py in zip(X[:10], y[:10], prediction[:10]):
                axes[0, 1].vlines(xi, py, yi, colors='gray', linestyles='--')
            axes[0, 1].set_title("Step 2: Errors (y - y_hat)")
            axes[0, 1].set_ylim(0, 15)

            # --- [3번 그래프] 초록색 손실 함수 지도와 현재 위치 ---
            axes[1, 0].cla()
            axes[1, 0].plot(w_range, l_range, 'g-', alpha=0.4) # 전체 지도
            axes[1, 0].plot(w_path, loss_path, 'ro-', markersize=5) # 이동 경로
            axes[1, 0].scatter(w, loss, color='red', s=100, edgecolors='black') # 현재 점
            axes[1, 0].set_title("Step 3: Loss Map & Current Point")
            axes[1, 0].set_xlabel("Weight (w)")
            axes[1, 0].set_ylabel("Loss")

            # --- [4번 그래프] 미분값(접선) 시각화 ---
            axes[1, 1].cla()
            axes[1, 1].plot(w_range, l_range, 'g-', alpha=0.2)
            # 접선(Tangent line) 그리기: 미분이 알려주는 방향
            t_w = np.linspace(w-0.5, w+0.5, 10)
            t_y = gradient * (t_w - w) + loss
            axes[1, 1].plot(t_w, t_y, 'k-', linewidth=2, label='Derivative (Slope)')
            axes[1, 1].scatter(w, loss, color='red', s=80)
            axes[1, 1].set_title("Step 4: Finding Direction via Derivative")
            axes[1, 1].set_xlabel("Weight (w)")
            axes[1, 1].legend()

            # 업데이트 및 한 걸음 이동
            plt.draw()
            plt.pause(0.2)
            w = w - 0.1 * gradient # 경사하강 실행
            
        time.sleep(2)

except KeyboardInterrupt:
    plt.ioff()
    print("종료되었습니다.")
```

```python
%matplotlib qt

import numpy as np
import matplotlib.pyplot as plt
import time

# 1. 4차 함수 형태의 가상 데이터 생성
np.random.seed(42)
X = np.linspace(-2, 2, 100).reshape(-1, 1)
# 실제 정답 곡선: y = 0.5x^4 - 3x^2 + x + 2
y = 0.5 * X**4 - 3 * X**2 + X + 2 + np.random.normal(0, 0.5, X.shape)

# 2. 4차 다항식 특징 행렬 만들기 [x^4, x^3, x^2, x, 1]
X_poly = np.c_[X**4, X**3, X**2, X, np.ones(X.shape)]

# 3. 그래프 설정
plt.ion()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

try:
    while True:
        # 가중치(theta) 랜덤 초기화 (5개의 파라미터)
        theta = np.random.randn(5, 1)
        learning_rate = 0.01
        loss_path = []
        
        for i in range(100):
            # 예측 및 오차 계산
            predictions = X_poly.dot(theta)
            errors = predictions - y
            loss = np.mean(errors**2)
            loss_path.append(loss)
            
            # 경사하강법 (미분값 계산)
            gradients = 2/100 * X_poly.T.dot(errors)
            theta = theta - learning_rate * gradients
            
            # --- 그래프 그리기 ---
            ax1.cla()
            ax1.scatter(X, y, color='red', alpha=0.5, label='Actual Data')
            ax1.plot(X, predictions, color='blue', linewidth=3, label='4th-order Model')
            ax1.set_title(f"Model Fitting (Step {i+1})")
            ax1.set_ylim(-10, 10)
            ax1.legend()

            ax2.cla()
            ax2.plot(loss_path, 'g-', linewidth=2)
            ax2.set_title("Loss over Time (How average error drops)")
            ax2.set_xlabel("Iteration")
            ax2.set_ylabel("Mean Squared Error")
            
            plt.draw()
            plt.pause(0.01) # 곡선은 복잡해서 조금 더 빨리 돌립니다.
            
        time.sleep(2)

except KeyboardInterrupt:
    plt.ioff()
    print("종료되었습니다.")
```

```python
%matplotlib qt

import numpy as np
import matplotlib.pyplot as plt
import time

# 1. 4차 함수 데이터 생성
np.random.seed(42)
X = np.linspace(-2, 2, 100)
y_true = 0.5 * X**4 - 3 * X**2 + 1 * X + 2 + np.random.normal(0, 0.4, X.shape)

# 2. 오차 지도(초록선) 미리 계산
w_range = np.linspace(-6, 0, 100)
l_range = [np.mean((0.5*X**4 + w*X**2 + 1*X + 2 - y_true)**2) for w in w_range]

# 3. 4분할 그래프 설정
plt.ion()
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
plt.subplots_adjust(hspace=0.3, wspace=0.2)

try:
    while True:
        w = np.random.uniform(-6, 0)
        w_path, loss_path = [], []
        
        # 30번의 스텝을 멈춤 없이 빠르게 진행
        for i in range(30):
            y_curr = 0.5 * X**4 + w * X**2 + 1 * X + 2
            error = y_curr - y_true
            loss = np.mean(error**2)
            gradient = np.mean(2 * error * X**2)
            
            w_path.append(w)
            loss_path.append(loss)

            # --- [1번: 모델 피팅] ---
            axes[0, 0].cla()
            axes[0, 0].scatter(X, y_true, color='red', alpha=0.3, s=20)
            axes[0, 0].plot(X, y_curr, color='blue', linewidth=3)
            axes[0, 0].set_title("1. Model Fitting")
            axes[0, 0].set_ylim(-10, 10)

            # --- [2번: 잔차 시각화] ---
            axes[0, 1].cla()
            axes[0, 1].scatter(X, y_true, color='red', alpha=0.3, s=20)
            axes[0, 1].plot(X, y_curr, color='blue', linewidth=2)
            for xi, yi, py in zip(X[::10], y_true[::10], y_curr[::10]):
                axes[0, 1].vlines(xi, py, yi, colors='gray', linestyles='--', alpha=0.5)
            axes[0, 1].set_title("2. Residuals (Errors)")
            axes[0, 1].set_ylim(-10, 10)

            # --- [3번: 오차 지도] ---
            axes[1, 0].cla()
            axes[1, 0].plot(w_range, l_range, 'g-', alpha=0.4)
            axes[1, 0].plot(w_path, loss_path, 'ro-', markersize=4, alpha=0.4)
            axes[1, 0].scatter(w, loss, color='red', s=80, edgecolors='black', zorder=5)
            axes[1, 0].set_title("3. Loss Map (Green Line)")

            # --- [4번: 미분 방향] ---
            axes[1, 1].cla()
            axes[1, 1].plot(w_range, l_range, 'g-', alpha=0.2)
            t_w = np.linspace(w-0.5, w+0.5, 10)
            t_y = gradient * (t_w - w) + loss
            axes[1, 1].plot(t_w, t_y, 'k-', linewidth=2)
            axes[1, 1].scatter(w, loss, color='red', s=60)
            axes[1, 1].set_title("4. Gradient Direction")

            plt.draw()
            plt.pause(0.05) # 속도감 있게 0.05초로 단축
            
            w = w - 0.1 * gradient 
            
        time.sleep(0.5) # 세션 사이 대기도 0.5초로 단축

except KeyboardInterrupt:
    plt.ioff()
    print("시뮬레이션을 종료합니다.")
```
