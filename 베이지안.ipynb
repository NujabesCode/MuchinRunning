```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. 강사님의 수식을 함수로 정의 (지형 만들기)
def objective_func(x, y):
    return x**2 - 20*y

# 2. 그래프를 그릴 데이터 생성
x_range = np.linspace(-10, 10, 100)
y_range = np.linspace(-15, 15, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = objective_func(X, Y)

# 3. 3차원 그래프 그리기
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 지형(Surface) 그리기
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# 4. 강사님이 찾은 'Best' 지점 표시 (20회 시도 결과)
# best: {'x': 2.0, 'y': 15.0}, loss: -296.0
best_x, best_y = 2.0, 15.0
best_z = objective_func(best_x, best_y)

ax.scatter(best_x, best_y, best_z, color='red', s=100, label='Best Point (20 evals)')

# 축 라벨 및 설정
ax.set_xlabel('X (hp.quniform)')
ax.set_ylabel('Y (hp.quniform)')
ax.set_zlabel('Loss (Return Value)')
ax.set_title('Bayesian Optimization Landscape')
plt.legend()
plt.show()
```
